## Distribution of Numerical Features
â—‹	Visualize the distribution of numerical features to identify patterns, skewness, and potential outliers.


```python
# Plotting histogram for Amount
plt.figure(figsize=(8, 6))
df['Amount'].hist(bins=10, edgecolor='black', color='skyblue')

# Adding title and labels
plt.title('Histogram of Amount', fontsize=16)
plt.xlabel('Amount', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show plot
plt.show()
```

```python
# Plotting histogram for Value
plt.figure(figsize=(8, 6))
df['Value'].hist(bins=10, edgecolor='black', color='skyblue')

# Adding title and labels
plt.title('Histogram of value', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show plot
plt.show()
```

```python
# Plotting histogram for PricingStrategy
plt.figure(figsize=(8, 6))
df['PricingStrategy'].hist(bins=10, edgecolor='black', color='skyblue')

# Adding title and labels
plt.title('Histogram of PricingStrategy', fontsize=16)
plt.xlabel('PricingStrategy', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show plot
plt.show()
```

```python
# Plotting the bar graph
plt.figure(figsize=(8, 6))
ax = value_counts.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.title('Counts of FraudResult', fontsize=16)
plt.xlabel('Fraud Result', fontsize=14)
plt.ylabel('Counts', fontsize=14)

# Adding count labels on top of each bar
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

# Display the plot
plt.show()
```

## Correlation Analysis

â—‹	Understanding the relationship between numerical features.


```python
# Computing the correlation matrix for the selected variables
correlation_matrix = df[['CountryCode', 'Amount', 'Value', 'PricingStrategy', 'FraudResult']].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Adding title
plt.title('Correlation Heatmap of Selected Variables', fontsize=16)

# Show the plot
plt.show()
```

## 	Distribution of Categorical Features
â—‹	Analyzing the distribution of categorical features provides insights into the frequency and variability of categories.


```python
# Count frequency of each category
frequency_df = df['TransactionId'].value_counts().reset_index()
frequency_df.columns = ['TransactionId', 'Frequency']

# Plotting bar plot using seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='TransactionId', y='Frequency', data=frequency_df, palette='Blues_d')

# Adding labels and title
plt.title('Bar Plot of TransactionId and Frequencies', fontsize=16)
plt.xlabel('TransactionId', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
```

```python
# Count frequency of each category
BatchId_df = df['BatchId'].value_counts().reset_index()
BatchId_df.columns = ['BatchId', 'Frequency']

# Plotting bar plot using seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='BatchId', y='Frequency', data=BatchId_df, palette='Blues_d')

# Adding labels and title
plt.title('Bar Plot of BatchId and Frequencies', fontsize=16)
plt.xlabel('BatchId', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
```

```python
# Count frequency of each category
StartTime_df = df['ProductCategory'].value_counts().reset_index()
StartTime_df.columns = ['ProductCategory', 'Frequency']

# Plotting bar plot using seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='ProductCategory', y='Frequency', data=StartTime_df, palette='Blues_d')

# Adding labels and title
plt.title('Bar Plot of TransactionStartTime and Frequencies', fontsize=16)
plt.xlabel('TransactionStartTime', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Show the plot
plt.show()
```

```python
# Count frequency of each category
ProductCategory_df = df['ProductCategory'].value_counts().reset_index()
ProductCategory_df.columns = ['ProductCategory', 'Frequency']

# Plotting bar plot using seaborn
plt.figure(figsize=(8, 6))
barplot = sns.barplot(x='ProductCategory', y='Frequency', data=ProductCategory_df, palette='Blues_d')

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45)

# Adding values on top of each bar
for index, row in ProductCategory_df.iterrows():
    barplot.text(index, row['Frequency'], round(row['Frequency'], 2), color='black', ha="center", va="bottom")

# Adding labels and title
plt.title('Bar Plot of ProductCategory and Frequencies', fontsize=16)
plt.xlabel('ProductCategory', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Adjust layout to avoid label overlap
plt.tight_layout()

# Show the plot
plt.show()
```
